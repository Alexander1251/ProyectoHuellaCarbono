```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde el CSV usando la ruta completa
data = pd.read_csv(r'C:\Users\jichv\SIC Clases y tareas\Proyecto Emisiones\datos_limpios.csv')

# Definir los pases de inters
paises_interes = [
    "Aruba", "Anguilla", "Antigua and Barbuda", "American Samoa", "Argentina",
    "Bahamas", "Barbados", "Belize", "Bermuda", "Bolivia", "Brazil", "Canada",
    "Chile", "Colombia", "Cuba", "Cayman Islands", "Dominican Republic",
    "Ecuador", "French Guiana", "Guatemala", "Guyana", "Honduras", "Jamaica",
    "Mexico", "Nicaragua", "Panama", "Peru", "Puerto Rico", "Suriname",
    "Trinidad and Tobago", "United States", "Venezuela", "Virgin Islands (British)",
    "Saint Vincent and the Grenadines"
]

# Filtrar los datos
filtered_data = data[data['pais_nombre'].isin(paises_interes)]

# Agrupar las emisiones anuales
annual_emissions = filtered_data.groupby('pais_nombre').sum(numeric_only=True).T

# Establecer un estilo de Seaborn
sns.set(style="whitegrid")

plt.figure(figsize=(14, 8))

# Colores de la paleta de Seaborn
colors = sns.color_palette("husl", len(annual_emissions.columns))

# Graficar cada pas
for i, country in enumerate(annual_emissions.columns):
    plt.plot(annual_emissions.index, annual_emissions[country], label=country, 
             alpha=0.8, linewidth=2, color=colors[i], marker='o', markersize=4)

# Ttulos y etiquetas
plt.title("Evolucin de las Emisiones por Pas en Amrica (1970-2023)", fontsize=16)
plt.xlabel("Ao", fontsize=14)
plt.ylabel("Emisiones Totales", fontsize=14)

# Mejorar la leyenda
plt.legend(loc='upper left', fontsize=10, ncol=2, frameon=True, title="Pases", title_fontsize='13')

# Rotar las etiquetas del eje x
plt.xticks(rotation=45)

# Agregar una cuadrcula
plt.grid(True, linestyle='--', alpha=0.7)

# Ajustar el diseo
plt.tight_layout()

# Mostrar la grfica
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Si ests en Jupyter Notebook, descomenta la siguiente lnea
# %matplotlib inline

# Cargar los datos desde el CSV usando la ruta completa
data = pd.read_csv('datos_limpios.csv')

# Transformar los datos para que sean ms fciles de graficar
data_melted = data.melt(id_vars=['pais_nombre', 'codigo_pais', 'region'], 
                         var_name='year', 
                         value_name='emissions')

# Convertir la columna 'year' a tipo numrico
data_melted['year'] = pd.to_numeric(data_melted['year'], errors='coerce')

# Convertir la columna 'emissions' a tipo numrico
data_melted['emissions'] = pd.to_numeric(data_melted['emissions'], errors='coerce')

# Filtrar los datos para eliminar filas con emisiones nulas o aos nulos
data_melted = data_melted.dropna(subset=['emissions', 'year'])

# Verificar los tipos de datos
print(data_melted.dtypes)

# Verificar que haya datos para graficar
print(data_melted.head())  # Muestra las primeras filas del DataFrame
print(data_melted.shape)    # Muestra la cantidad de filas y columnas

# Verificar si hay datos para graficar
if data_melted.empty:
    print("No hay datos para graficar.")
else:
    # Crear la grfica
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=data_melted, x='year', y='emissions', hue='region', marker='o')

    # Configurar el ttulo y las etiquetas
    plt.title('Evolucin de las Emisiones por Regin a lo Largo de los Aos', fontsize=16)
    plt.xlabel('Ao', fontsize=14)
    plt.ylabel('Emisiones', fontsize=14)
    plt.legend(title='Regin', fontsize=12)
    plt.grid()

    # Mostrar la grfica
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



ruta_archivo = "datos_limpios.csv"

# Cargar los datos
data = pd.read_csv(ruta_archivo)



# Preparar datos para la grfica de lneas
emisiones_por_anio = data.iloc[:, 4:].sum(numeric_only=True)

# Preparar datos para la grfica de barras apiladas (ltimos 5 aos)
ultimos_5_anios = ['2019', '2020', '2021', '2022', '2023']
emisiones_por_region = (
    data.groupby('region')[ultimos_5_anios]
    .sum(numeric_only=True)
    .reset_index()
)

# Configurar estilo de las grficas
sns.set_theme(style="whitegrid")

# Grfica de lneas
plt.figure(figsize=(12, 6))
sns.lineplot(x=emisiones_por_anio.index, y=emisiones_por_anio.values, marker="o")
plt.title("Evolucin de las emisiones globales (1970-2023)", fontsize=16)
plt.xlabel("Ao", fontsize=12)
plt.ylabel("Emisiones totales", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Grfica de barras apiladas
emisiones_por_region_melted = emisiones_por_region.melt(id_vars="region", var_name="Ao", value_name="Emisiones")
plt.figure(figsize=(12, 8))
sns.barplot(
    data=emisiones_por_region_melted,
    x="Ao",
    y="Emisiones",
    hue="region",
    palette="tab20"
)
plt.title("Emisiones por regin en los ltimos 5 aos (2019-2023)", fontsize=16)
plt.xlabel("Ao", fontsize=12)
plt.ylabel("Emisiones", fontsize=12)
plt.legend(title="Regin", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
